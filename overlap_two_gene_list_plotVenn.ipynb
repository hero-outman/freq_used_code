```python
# check if list too long, make it 3 genes one row
def make_list_genes_to_str(list_to_check):
    length_of_list = len(list_to_check)

    textStr = ''

    if length_of_list > 10:
        for i,l in enumerate(list_to_check):
            textStr = textStr + l
            gene_order = i + 1
            if gene_order == length_of_list - 1:
                pass
            if gene_order % 3 != 0 and i != length_of_list - 1:
                textStr = textStr + ', '
            if gene_order % 3 == 0 :
                textStr = textStr + '\n'
    if length_of_list <= 10:
        textStr = '\n'.join(list_to_check)

    return textStr
```
