```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
os.chdir('/Users/jplab/Desktop/DAILY_CODE_DATA/2022-7/data/7-8_TFbound_change_withFOXA1KD/')
```

```python
deseq2Norm_counts = '/Users/jplab/Desktop/TOBIAS/bindetect_shANDnc/jaspar/bindetect_results_shANDnc.xlsx'

df_deseq2Norm_counts = pd.read_excel(deseq2Norm_counts,header=0)
```

```python
keep = ['name','shB14-ncB14','shDMSO-ncDMSO','shTGF-ncTGF']
df_change = df_deseq2Norm_counts[keep]
df_change.index = df_change['name']
```

```python
color_list = [
    '#FFBF00', 
    # '#FF7F50', 
    '#DE3163', 
    # '#9FE2BF', 
    # '#40E0D0', 
    '#6495ED'
    ]

fig_title = 'changes of TFs bound number after FOXA1 knock down between B1.14, DMSO and TGF-beta conditions'
fig_xlabel = 'TF name'
fig_ylabel = 'change of TFs bound number \nafter FOXA1 knock down'

fig_weight = len(df_change.index) / 2
fig_height = 12
ax= df_change.plot(
    kind='bar', 
    stacked=False,
    figsize=(fig_weight,fig_height),
    color=color_list,
    width=0.85,
    zorder=1
    )
    
plt.legend(prop={'size': 20})
plt.title(fig_title,size=42)
plt.xlabel(fig_xlabel,fontsize=30)
plt.xticks(rotation=90, horizontalalignment="center",fontsize=20)
plt.ylabel(fig_ylabel,fontsize=30)
plt.yticks(fontsize=20)
plt.grid(axis='y', linestyle='--', linewidth=0.5, color='grey', alpha=0.5, zorder=0)
```
