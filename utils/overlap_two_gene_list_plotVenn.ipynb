# a method to arange contents of venn circle

```python
# check if list too long, make it 3 genes one row
def make_list_genes_to_str(list_to_check,max_genes_to_split=10,each_row_num=3):
    length_of_list = len(list_to_check)

    textStr = ''

    if length_of_list > max_genes_to_split:
        for i,l in enumerate(list_to_check):
            textStr = textStr + l
            gene_order = i + 1
            if gene_order == length_of_list - 1:
                pass
            if gene_order % each_row_num != 0 and i != length_of_list - 1:
                textStr = textStr + ', '
            if gene_order % each_row_num == 0 :
                textStr = textStr + '\n'
    if length_of_list <= 10:
        textStr = '\n'.join(list_to_check)

    return textStr
```
