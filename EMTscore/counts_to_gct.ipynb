````python
import os
import pandas as pd

os.chdir('/Users/jplab/Desktop/DAILY_CODE_DATA/2022-4/data/4-6_EMTScore/gct/')
````

````python
# deseq2 normed data with gene name
deseq2_norm_counts = '/Users/jplab/Desktop/DAILY_CODE_DATA/2022-4/code/4-6_EMTScore/counts/mRNA_seq_FOXA1_KD_NC_2022221.tsv'
df_deseq2_norm_counts = pd.read_csv(deseq2_norm_counts, sep='\t',header=0)
df_deseq2_norm_counts.drop_duplicates(subset=['geneid'], inplace=True)


df_deseq2_norm_counts
````

````python
columns = df_deseq2_norm_counts.columns.to_list()
start_offset = columns.index('geneid')
end_offset = columns.index('baseMean')
sample_names = columns[start_offset+1:end_offset]
sample_numbers = len(sample_names)

# df_deseq2_norm_counts['baseMean'] = df_deseq2_norm_counts.mean(axis=1,numeric_only=True)

third_line = 'Name\tDescription\t' + '\t'.join(sample_names)

df_deseq2_norm_counts['Description'] = 'na'

keep_order = ['external_gene_name','Description']+sample_names
df_keep = df_deseq2_norm_counts[keep_order]
df_keep['external_gene_name'] = df_keep['external_gene_name'].str.upper()
df_keep['Description'] = df_keep['external_gene_name'].str.upper()
df_keep.rename(columns={'external_gene_name':'NAME'}, inplace=True)

df_keep.drop_duplicates(subset='NAME', keep='first', inplace=True)
df_keep.dropna(subset=['NAME'], inplace=True)

df_keep = df_keep[df_keep.mean(axis=1,numeric_only=True) > 50]

# how many genes are in the dataframe
gene_numbers = df_keep.shape[0]
gene_numbers

first_line = '#1.2'
# defalut 4, and will change by the number of actual columns
sample_numbers = 12

second_line = str(gene_numbers) + '\t' + str(sample_numbers)
second_line

````

````python
save_as = 'mRNA_seq_FOXA1_KD_NC_2022221.gct'

with open(save_as, 'w') as f:
    f.write(first_line + '\n')
    f.write(second_line + '\n')
    f.write(third_line + '\n')
    for index, row in df_keep.iterrows():
        f.write(row['NAME'] + '\t' + row['Description'] + '\t')
        for i,s in enumerate(sample_names):
            f.write(str(row[s]))
            if i != sample_numbers-1:
                f.write('\t')
        f.write('\n')
````
